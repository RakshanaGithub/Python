# WEB-SCRAPING ON AMAZON BESTSELLERS IN BEAUTY

```python
# IMPORTING NECCESARY PACKAGES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
import time
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
no_pages = 3
# function get_data is defined 
def get_data(pageNo):  
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Edg/117.0.2045.31"}

    r = requests.get('https://www.amazon.in/gp/bestsellers/beauty/ref=zg_bs_beauty_sm'+str(pageNo)+'?ie=UTF8&pg='+str(pageNo), headers=headers)#, proxies=proxies)
    content = r.content
    soup = BeautifulSoup(content) #The content is extracted and parsed using beautifulsoup library
    

    alls = []
    for d in soup.findAll('div', attrs={'class':'a-cardui _cDEzb_grid-cell_1uMOS expandableGrid p13n-grid-content'}):
        #print(d)
        name = d.find('div', attrs={'class':'_cDEzb_p13n-sc-css-line-clamp-3_g3dy1'})
        #print(n[0]['alt'])
        rating = d.find('span', attrs={'class':'a-icon-alt'})
        users_rated = d.find('span', attrs={'class':'a-size-small'})
        price = d.find('span', attrs={'class':'_cDEzb_p13n-sc-price_3mJ9Z'})
        all1=[]

        if name is not None:
            all1.append(name.text)
        else:
            all1.append("unknown-product")

        if rating is not None:
            all1.append(rating.text)
        else:
            all1.append('-1')

        if users_rated is not None:
            all1.append(users_rated.text)
        else:
            all1.append('0')     

        if price is not None:
            all1.append(price.text)
        else:
            all1.append('0')
        alls.append(all1)    
    return alls
```

```python
results = []
for i in range(1, no_pages+1):
    results.append(get_data(i))
flatten = lambda l: [item for sublist in l for item in sublist] #takes a list of lists and flattens it into a single list
df = pd.DataFrame(flatten(results),columns=['Name','Rating','Customers_Rated', 'Price']) 
#The list is converted into a pandas dataframe and this is turned into a csv 
df.to_csv('amazon_products.csv', index=False, encoding='utf-8')
```
